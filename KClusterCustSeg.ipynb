```python
#importing needed libraries
import random
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_samples, silhouette_score
%matplotlib inline
```

```python
#setting random seed for data
np.random.seed(4)
```

```python
#making feature matrix and response vector
X, y = make_blobs(n_samples=5000, centers=[[4,4],[-2,-1], [2,-3],[1,1]], cluster_std=0.9)
```

```python
#disply plot of random data
plt.scatter(X[:, 0], X[:, 1], marker='.')
plt.show()
```

```python
#Setting up K-means Clustering model with 4 cluster groups and k-means++
kmeans = KMeans(n_clusters=4, init='k-means++', n_init=12, max_iter=300, tol=0.0001, copy_x=True, algorithm='auto')
```

```python
#fiting kmeans cluster model to data
kmeans.fit(X)
```

```python
#gtting model labels
k_mean_labels = kmeans.labels_
k_mean_labels
```

```python
#getting model cluster centers
k_mean_cluster_centers = kmeans.cluster_centers_
k_mean_cluster_centers
```

```python
#ploting kmeans cluster model of customer segmens
fig, ax = plt.subplots()
ax.scatter(X[:, 0], X[:, 1], c=k_mean_labels, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
ax.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
ax.set_title('K-Means Clustering Customer Segments')
ax.set_xlabel('Age')
ax.set_ylabel('Income')
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```
