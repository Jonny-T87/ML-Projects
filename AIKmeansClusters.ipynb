```python
import pickle
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.cluster import KMeans, MeanShift
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import seaborn as sns; sns.set()
import sklearn.metrics as sm
```

```python
data = np.loadtxt('data.csv', delimiter=',')

X,y = data[:,:-1], data[:,-1]
```

```python
#Split into Training and Test
num_train = int(0.8*len(X))
X_train, y_train = X[:num_train], y[:num_train]
X_test, y_test = X[num_train:], y[num_train:]
```

```python
#regressor model
regressor = linear_model.LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#predict with testing data
y_test_pred = regressor.predict(X_test)
```

```python
#plotting output
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_test_pred, color='blue', linewidth=3)
plt.yticks(())
plt.xticks(())
plt.show()
```

```python
#Importing data
from sklearn.datasets import make_blobs
```

```python
#making X and y datasets
X, y_true = make_blobs(n_samples=500, centers=4,
                  random_state=4, cluster_std=0.40)
```

```python
#taking a look at the data
print(X)
print(y_true)
```

```python
#ploting data in scatter plot
plt.scatter(X[:, 0], X[:, 1], s=50, alpha= 0.5)
plt.show()
```

```python
#setting k-means clusters
kmeans = KMeans(n_clusters=4, random_state=4)
```

```python
#fitting model to data
kmeans.fit(X)
```

```python
#predicting with model
y_kmeans = kmeans.predict(X)
```

```python
#ploting predicitions
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, alpha=0.5, cmap='viridis')
plt.show()
```

```python
#finding kmeans model centerpoint
centers = kmeans.cluster_centers_
print(centers)
```

```python
#ploting center points
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```

```python
#plotting clusters w/ center points
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, alpha=0.5, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```
